# Multi-class Chest X-ray Disease Classification Using Transfer Learning — Selected Classes Organizer

## Overview
This notebook organizes a subset of the **NIH Chest X-ray dataset** into per-class folders.
It reads a metadata CSV containing image filenames and their associated labels
(`Finding Labels`), then copies each image from the raw dataset directory into a
new destination directory, structured as one folder per class.

**Pipeline:**
1. Import libraries
2. Define dataset paths
3. Load metadata
4. Create class folders
5. Copy images into their class folders
6. Count copied images

## 1. Import Libraries

Standard libraries for file handling and metadata processing.

```python
import os
import shutil
import pandas as pd
```

## 2. Define Dataset Paths

All input/output paths are defined here in one place for easy configuration.

- `METADATA_CSV`: CSV file with image names and labels
- `SOURCE_DIR`: root directory containing the raw NIH dataset (subfolders `images_001` ... `images_012`)
- `DESTINATION_DIR`: output directory where images will be organized by class

```python
METADATA_CSV = r"..\data\selected_images_7class.csv"
SOURCE_DIR = r"..\data\archive"
DESTINATION_DIR = r"..\data\NIH_selected_7class"
NUM_IMAGE_FOLDERS = 12  # dataset is split into images_001 ... images_012
```

## 3. Load Metadata

Load the CSV file containing image filenames and their corresponding class labels.

```python
df = pd.read_csv(METADATA_CSV)
classes = df["Finding Labels"].unique()

print(f"Total records: {len(df)}")
print(f"Number of classes: {len(classes)}")
```

## 4. Create Class Folders

Create the destination directory and one subfolder per class.

```python
os.makedirs(DESTINATION_DIR, exist_ok=True)

for cls in classes:
    os.makedirs(os.path.join(DESTINATION_DIR, cls), exist_ok=True)

print("Class folders created successfully.")
```

## 5. Copy Images

For each row in the metadata, search for the image inside the numbered
`images_XXX` subfolders and copy it into its corresponding class folder.

```python
not_found = []

for _, row in df.iterrows():
    image_name = row["Image Index"]
    label = row["Finding Labels"]

    found = False
    for i in range(1, NUM_IMAGE_FOLDERS + 1):
        folder = f"images_{i:03d}"
        image_path = os.path.join(SOURCE_DIR, folder, "images", image_name)

        if os.path.exists(image_path):
            dest_path = os.path.join(DESTINATION_DIR, label, image_name)
            shutil.copy2(image_path, dest_path)
            found = True
            break

    if not found:
        not_found.append(image_name)

print("Copy finished.")
if not_found:
    print(f"{len(not_found)} image(s) were not found:")
    print(not_found)
```

## 6. Count Copied Images

Verify the operation by counting the total number of images copied into
the destination directory.

```python
total = 0
for _, _, files in os.walk(DESTINATION_DIR):
    total += len(files)

print("Total images copied:", total)
```
